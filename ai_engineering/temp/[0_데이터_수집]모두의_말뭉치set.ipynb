```python
# "일기" 단어가 포함된 행을 추출하는 함수 정의
def extract_diary_rows(diary):
    return diary[diary['category'].str.contains('일기')]
```

```python
# 구두점 기반 토크나이저 함수 정의
import re

def count_sentences(text):
    sentences = re.split(r'[.!?]', text) # 정규표현식을 사용하여 문장을 구두점을 기준으로 분리
    sentences = [sentence.strip() for sentence in sentences if sentence.strip()]
    return len(sentences)
```

```python
diary = pd.read_csv("/content/drive/MyDrive/0000/Class/졸프/Dataset/모두의말뭉치/NIKL_NP.csv")
diary = diary.rename(columns={'file_id':'id','sentence': 'text'})

# "일기" 단어가 포함된 행 추출
diary = extract_diary_rows(diary)
diary = diary[diary['author_sex']=='F']
diary = diary[diary["author_occupation"].isin(["직장인/전문직", "전업주부", "자영업자", "무직"])] # 직업 선택
#diary = diary[diary["author_occupation"].isin(["전업주부"])] # 직업 선택
diary = diary[(diary["author_age"] >= 25) & (diary["author_age"] <= 40)] # author_age가 25세 이상 45세 이하인 행만 선택

diary = diary[['id','title','text_subclass','text']] # 컬럼 정리

# 텍스트 병합(조건 : 문장 길이 2 이상)
diary_text = diary.groupby('id')['text'].apply(lambda x: ' '.join(x)).reset_index() # 아이디를 기준으로 그룹화하여 텍스트를 합치기
diary_text['text_cnt'] = diary_text['text'].apply(count_sentences) # 텍스트 열에 적용하여 문장 개수를 계산하고 새로운 열로 추가
diary_text = diary_text[(diary_text['text_cnt'] >= 2)&(diary_text['text_cnt']<=36)] # 문장 개수가 2 이상인 행만 선택하여 새로운 데이터프레임 생성
diary_text = diary_text[['text']]

diary_text
```
